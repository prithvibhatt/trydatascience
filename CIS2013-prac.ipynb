```python
# Lecture 10
#Function to calculate bill
def calc_bill(price, qty):
    lineitem= price*qty
    return lineitem
    

#Function to calculate VAT
def calc_VAT(amt, VAT):
    final= amt + (amt*VAT)
    return final

#Calculate bill
num = input("Enter how many items")
num= int(num)
i=0
tot=0
while i<num:
    price = input("Enter price of item " + str(i))
    price = float(price)
    qty = input("Enter quantity of item " + str(i))
    qty = float(qty)
    ln = calc_bill (price,qty)
    tot= tot+ln
    i=i+1
print (tot)
```

```python
# Lecture 11 to 15

#Function to calculate grade based on agiven mark
def calc_grade(mark):
    if(mark>=90):
      Grade = "A"
    elif(mark>=87 and mark<=89):
      Grade="A-"
    elif(mark>=84 and mark<=86):
      Grade="B+"
    elif(mark>=80 and mark<=83):
      Grade = "B"
    elif(mark>=77 and mark<=79):
      Grade = "B-"
    elif(mark>=74 and mark<=76):
      Grade = "C+"
    elif(mark>=70 and mark<=73):
      Grade = "C"
    elif(mark>=67 and mark<=69):
      Grade = "C-"
    elif(mark>=64 and mark<=66):
      Grade = "D+"
    elif(mark>=60 and mark<=63):
      Grade = "D"
    else:
      Grade = "F"
    return Grade

#Function to calculate GPA based on a given mark
def calc_GPA(mark):
    GPA = (mark/100)*4 
    return GPA

# calculating grade using function
mk = input("Enter the mark you want the grade for: ")
mk= float(mk)
grd= calc_grade(mk)
print ("Your grade is: " + str(grd)) 

# calculating GPA using function
mk = input("Enter the mark you want the GPA for: ")
mk= float(mk)
gpa= calc_GPA(mk)
print ("Your GPA is: " + str(gpa)) 

# Calculating GPA of many courses for the semester
numCourses = 3
tot = 0
#Creating an array
marks = [0 for x in range(numCourses)] 
for i in range(0,numCourses):
    mk = input("Enter the mark for the course (out of 100): ")
    marks[i] = float(mk)
    tot = tot + marks[i]
    avg = tot/numCourses
    GPA = tot/(numCourses*100)*4
print ("The total marks for the semester is: " + str(tot)) 
print ("The average marks for the semester is: " + str(avg)) 
print ("The GPA for the semester is: " + str(GPA)) 


# testing for a number to be positive or not
n1 = input("Enter a number to test for positive or negative: ")
n1= float(n1)

if n1 > 0:
  print("the number is  positive")
elif n1 < 0:
  print("the number is negative")
else:
  print("the number is 0")
  
  
# Testing for the greater of 2 numbers
num1 = 100
num2 = 10
if num1 > num2:
  print(str(num1) + "is larger")
elif num1 < num2:
  print(str(num2) + "is larger")
else:
    print(str(num1) + " and " + str(num2) + " are equal")
  
```
